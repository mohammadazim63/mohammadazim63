### Onehotencoding with many categories in a feature

```python
import pandas as pd
```

```python
df = pd.read_csv('mercedes.csv',usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
#Now using iterator i function we can find unique values in each categorical features
for i in df.columns:
    print(df[i].unique())
```

```python
# Now find the length of unique categorical features
for i in df.columns:
    print(len(df[i].unique()))
```

```python
# Next take first top 10 records to perform one hot encoding
df['X1'].value_counts().sort_values(ascending=False).head(10)
```

```python
# In this step find index of all 10 variable
df['X1'].value_counts().sort_values(ascending=False).head(10).index
```

```python
# In this step make list of those 10 variables
lst_10=df['X1'].value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
```

```python
lst_10
```

```python
import numpy as np
for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```
